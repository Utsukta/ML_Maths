1. The Calculus of Infinitesimals

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x= np.linspace(-10,10,1000)
```

If y=x^2+2x+2:

```python
y=x**2+2*x+2
```

```python
fig,ax=plt.subplots()
ax.plot(x,y)
```

```python
fig,ax=plt.subplots()
ax.set_xlim([-2,0])
ax.set_ylim([0,2])
ax.plot(x,y)
```

```python
fig,ax=plt.subplots()
ax.set_xlim([-1.25,-0.5])
ax.set_ylim([0.5,1.5])
ax.plot(x,y)
```

```python
fig,ax=plt.subplots()
ax.set_xlim([-1.1,-0.9])
ax.set_ylim([0.9,1.1])
ax.plot(x,y)
```

```python
fig,ax=plt.subplots()
ax.set_xlim([-1.01,-0.99])
ax.set_ylim([0.99,1.01])
ax.plot(x,y)
```

PARTIAL DERIVATIVES

```python
import math
def cylinder_vol(x,y):
    return math.pi*x**2*y
```

```python
import torch
r=torch.tensor(3.0).requires_grad_()
```

```python
l=torch.tensor(4.0).requires_grad_()
```

```python
v=cylinder_vol(r,l)
v
```

```python
v.backward()
```
